Question 4: Write a python program to scrap book name, author name, genre and book review of any 5 books from
‘www.bookpage.com’

```python
page=requests.get("https://bookpage.com/features")
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
web=soup.find_all("div", class_="flex-article-content")
web
```

Book Name

```python
books=[]

for i in web:
    for j in i.find_all("h4"):
        books.append(j.text.replace("\n",""))
        
books
```

Author, Genere, Review

```python
author=[]

for i in web:
    for j in i.find_all("p"):
        author.append(j.text.replace("\n",""))
        
author
```

ii) Top 10 ODI Batsmen in men along with the records of their team and rating.
iii) Top 10 ODI bowlers along with the records of their team andrating.
